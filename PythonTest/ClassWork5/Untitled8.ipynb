```python
# 2
np.random.seed(20)
ar_poiss = np.random.poisson(15, size=1000)

x = ar_poiss[:len(ar_poiss)//2]
y = ar_poiss[len(ar_poiss)//2:]

x1 = 20
y1 = 20
distribution, x1, y1 = np.histogram2d(x, y, bins=(x1, y1))
cs = plt.matshow(distribution)
plt.colorbar(cs)
plt.show()
```
